```python
secret = 33
attempts = 0

while True:
    guess = int(input("Guess : "))
    attempts += 1

    if guess > secret:
        print("Too high")
    elif guess < secret:
        print("Too low")
    else:
        print("Correct!")
        print("Attempts:", attempts)
        break
```

```python
shopping = []

while True:
    print("1- Add item")
    print("2- Remove item")
    print("3- Show sorted list")
    print("4- Exit")

    choice = input("Choose: ")

    if choice == "1":
        item = input("Enter item: ")
        shopping.append(item)

    elif choice == "2":
        item = input("Enter item to remove: ")
        if item in shopping:
            shopping.remove(item)
        else:
            print("Item not found")

    elif choice == "3":
        print(sorted(shopping))

    elif choice == "4":
        break

    else:
        print("Invalid choice")
```
